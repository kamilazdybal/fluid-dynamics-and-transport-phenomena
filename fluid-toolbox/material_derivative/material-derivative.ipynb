```python
import numpy as np
import matplotlib.pyplot as plt
import cmcrameri as cmc
import scipy
cmap = cmc.cm.oslo
```

```python
np.random.seed(seed=20)
```

```python
H = 200
W = 200
```

```python
x = np.linspace(0,W,W)
y = np.linspace(0,H,H)
```

```python
x_grid, y_grid = np.meshgrid(x, y)
```

```python
field_t1 = np.random.rand(H, W) * 1000
for _ in range(0, 10):
    field_t1 = scipy.ndimage.gaussian_filter(field_t1, 6)
```

```python
for _ in range(0, 4):
    field_t2 = scipy.ndimage.gaussian_filter(field_t1, 6)
```

```python
field_t2 = field_t2 + (x_grid**2)/2000
```

```python
plt.imshow(field_t1, cmap=cmap)
plt.xticks([])
plt.yticks([]);
cb = plt.colorbar(aspect=4)
cb.outline.set_visible(False)
cb.set_ticks([])
plt.savefig('t1.svg', dpi=300, bbox_inches='tight')
```

```python
plt.imshow(field_t2, cmap=cmap)
plt.xticks([])
plt.yticks([]);
# plt.colorbar()
plt.savefig('t2.svg', dpi=300, bbox_inches='tight')
```

```python
velocity_u = np.cos(x_grid/200) * np.sin(y_grid/200)
velocity_v = field_t1/1000
```

```python
np.max(field_t2)
```

```python
np.max(velocity_u)
```

```python
fig = plt.figure(figsize=(5,5))
plt.streamplot(x_grid, y_grid, velocity_u, velocity_v, density=0.6, color='k')#, color='#FF0080')
plt.xticks([])
plt.yticks([])
plt.savefig('vector-field.svg', dpi=300, bbox_inches='tight', transparent=True)
```
